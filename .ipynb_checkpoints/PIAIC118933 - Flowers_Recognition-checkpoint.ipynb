# Loading Data

```python
# using Pillow library to load data 

import numpy as np
import os
import os.path
import PIL
from PIL import Image

# let us make a function that can create dataset containing images and thier labels

def flower_dataset(image_folder):
    
    image_data=[]
    image_label=[]
    for folders in os.listdir(image_folder):
        for photo in os.listdir(os.path.join(image_folder, folders)):
       
            image_path= os.path.join(image_folder, folders,  photo)
            image= Image.open(image_path)            
            image = np.array(Image.open(image_path))
            image = image.astype('float32')
            image_data.append(image)
            image_label.append(folders)
    return image_data , image_label
image_data, image_label = flower_dataset(r'flowers')
```

```python
# let us check the data at 0 index
image_data[0], image_label[0]
```

# Resizing and converting to grayscale

The following steps for resizing and converting to grayscale could not be fullfilled <br> inside flower_dataset function, so I am doing the following for each individual sub folder

```python
folders = r'flowers/daisy' 
# I changed the sub directory name like sunflower and others and run this cell again and again for resizing
for file in os.listdir(folders):
    f_img = folders+"/"+file
    img = Image.open(f_img)
    img = img.resize((200,200))
    img = img.convert('L') # converting to grayscale
    img.save(f_img)
```

# Labels encoding with mapping

```python
label_dict={k: v for v, k in enumerate(np.unique(image_label))}
label_dict
```

```python
label_val = [label_dict[image_label[i]] for i in range(len(image_label))]
```

# Distributing data into data and labels

```python
import numpy as np
x = np.array(image_data, np.float32)
```

```python
y = np.array(list(map(int,label_val)), np.float32)
```

# Splitting data with train_test_split

```python
# distribution into X and y & spliting data
from sklearn.model_selection import train_test_split
train_ratio = .6 # given ratio is 60% as training
test_ratio = .4 # given 40 % as test

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = test_ratio) # as test size is 40%
len(y_train), len(y_test)
```

# Preparing image data

```python
x_train = np.array(x_train).reshape((2593, 200 * 200))
x_train = x_train.astype('float32') / 255
x_test = np.array(x_test).reshape((1729, 200 * 200))
x_test = x_test.astype('float32') / 255
```

# Preparing the labels

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

# Creating model

```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.optimizers import *

model = Sequential()
#tf.keras.backend.set_floatx('float64')
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dense(5, activation='softmax')) # as 5 labels are there 

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=180)
```

# Evaluation

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('test_acc:', test_acc)
```
